```python
#ploting count vs size
ax = sns.countplot(x = 'Size', data = df)
```

```python
#with numbers
ax = sns.countplot(x = 'Size', data = df)
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
#table view for size and quntity sum, group by function is used to group columns
df.groupby(['Size'], as_index = False)['Quantity'].sum().sort_values(by='Quantity', ascending = False)
```

```python
#plot for Size vs Quantity
S_Qty=df.groupby(['Size'], as_index = False)['Quantity'].sum().sort_values(by='Quantity', ascending = False)
ax=sns.barplot(x= 'Size', y= 'Quantity', data = S_Qty)
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
#checking shipping efficiency
sns.countplot(data=df, x='Courier Status', hue= 'Status')
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df, x='Courier Status', hue= 'Status')
plt.show()
```

```python
#histogram for size
df['Size'].hist()
```

```python
# Category count bar plot
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=45)
plt.show()
```

```python
B2B_check = df["B2B"].value_counts()
plt.pie(B2B_check, labels=B2B_check, autopct='%1.0f%%')
plt.show()
```

```python
#B2B checks 
B2B_check = df["B2B"].value_counts()
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot for Size and Category
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
```

```python
# State wise sale
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution State Wise')
```

```python
#Top 12 states, sales wise
top_12 = df['ship-state'].value_counts().head(12)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_12.index)], x='ship-state', order = top_12.index)
plt.xticks(rotation=48)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution State Wise')
```
